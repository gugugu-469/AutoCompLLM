```python
import os
import jsonlines
import json
```

```python
# Each lien format
# Training time, Inf time, p, r, f1, vd, vt, vi 
# Training time, Inf time, acc, vd, vt, vi 
all_results = [
    {
        "res":'''
''',
    "out_name":"CMeIE-qwen3"
},
{
"res":'''
''',
"out_name":"CMeIE-glm4"
},
{
"res":'''
''',
"out_name":"CMeIE-llama3"
},
{
"res":'''
''',
"out_name":"CMeIE-intern3"
}
]
```

```python
top_num = 4
top_num_2 = 3
```

```python
for item in all_results:
    s = item['res']
    out_name = item['out_name']
    print(out_name)
    s = s.strip()
    best_str_list = ''
    other_str_list = []

    # Create a list to store all results with their metrics
    results = []
    
    if 'Common' in out_name:
        for item in s.strip().split('\n'):
            tra_time, inf_time, acc, vd, vt, vi = item.strip().split('\t')
            other_str_list.append('{}\t{}\t{}'.format(vd, vt, vi))
            acc = float(acc)
            total_time = float(tra_time) + float(inf_time)
            results.append({
                'acc': acc,
                'time': total_time,
                'vd': vd,
                'vt': vt,
                'vi': vi
            })
        
        # Sort by accuracy descending, then by time ascending
        results.sort(key=lambda x: (-x['acc'], x['time']))
        # Get top 5 by accuracy
        top5 = results[:top_num_2]
        print(top5)
        # From top5, get the one with smallest time
        if top5:
            best = min(top5, key=lambda x: x['time'])
            print(best)
            best_str_list = '{}\t{}\t{}'.format(best['vd'], best['vt'], best['vi'])
            
    else:
        for item in s.strip().split('\n'):
            tra_time, inf_time, p, r, f1, vd, vt, vi = item.strip().split('\t')
            other_str_list.append('{}\t{}\t{}'.format(vd, vt, vi))
            f1 = float(f1)
            total_time = float(tra_time) + float(inf_time)
            results.append({
                'f1': f1,
                'time': total_time,
                'vd': vd,
                'vt': vt,
                'vi': vi
            })
        
        # Sort by f1 descending, then by time ascending
        results.sort(key=lambda x: (-x['f1'], x['time']))
        # Get top 5 by f1
        top5 = results[:top_num]
        print(top5)
        # From top5, get the one with smallest time
        if top5:
            best = min(top5, key=lambda x: x['time'])
            print(best)
            best_str_list = '{}\t{}\t{}'.format(best['vd'], best['vt'], best['vi'])

    with open('./f1_effi/{}.json'.format(out_name), 'w') as f:
        json.dump({
            'other_str_list': other_str_list,
            'best_str_list': best_str_list,
        }, f, ensure_ascii=False, indent=2)
```
